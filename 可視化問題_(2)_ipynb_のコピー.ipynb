## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

```python
data1 = pd.read_csv("https://raw.githubusercontent.com/ekikan/Basic-Python/master/colab/quizzes/matplotlib_quiz/hist_data.csv")
plt.hist(data1['aデータ'],bins=30,alpha=0.5,color="orange",label="aデータ")
plt.hist(data1['bデータ'],bins=30,alpha=0.5,color="skyblue",label="bデータ")
plt.xlabel("温度(℃)")
plt.ylabel("度数")
plt.title("aデータとbデータの温度に関する度数分布")
plt.legend()
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

```python
data2 = pd.read_csv("https://raw.githubusercontent.com/ekikan/Basic-Python/master/colab/quizzes/matplotlib_quiz/scatter_data.csv")
data2_class0 = data2[data2["class"]==0]
data2_class1 = data2[data2["class"]==1]
data2_class2 = data2[data2["class"]==2]
plt.scatter(data2_class0["length (cm)"],data2_class0["width (cm)"],color="red",label="class0")
plt.scatter(data2_class1["length (cm)"],data2_class1["width (cm)"],color="blue",label="class1")
plt.scatter(data2_class2["length (cm)"],data2_class2["width (cm)"],color="green",label="class2")
plt.xlabel("length (cm)")
plt.ylabel("width (cm)")
plt.title("クラス0～2のlengthとwidthに関する散布図")
plt.grid(True)
plt.legend()
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

```python
data3=pd.read_csv("https://raw.githubusercontent.com/ekikan/Basic-Python/master/colab/quizzes/matplotlib_quiz/line_data.csv")
plt.figure(figsize=(6,9))
plt.subplot(2,1,1)
plt.plot(data3["期間"],data3["件数"],marker="o")
plt.xlabel("期間")
plt.ylabel("件数")
plt.title("期間ごとの件数の推移")
plt.xticks(rotation=45)
plt.grid(True)
plt.subplot(2,1,2)
plt.bar(data3["期間"],data3["費用"])
plt.xlabel("期間")
plt.ylabel("費用(円)")
plt.title("期間ごとの費用の推移")
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```
